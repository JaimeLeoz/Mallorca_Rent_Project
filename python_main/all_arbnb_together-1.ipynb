```python
# ahora hacemos formula para calcular el cambio de moneda por año. Ya que están los precios en dolares.
```

```python
dictio = {'2016':1.16,
         '2017':1.13,
         '2018':1.18,
         '2019':1.11,
         '2020':1.14,
         '2021':1.18,
         '2022':1}
```

```python
clean_data['price_night_euros'] = [clean_data.price[i] * dictio[clean_data.last_scraped_year[i]] for i in range(len(clean_data))]
```

```python
# we need to change the column sqm into a integer befor calculating anything
clean_data["sq_meters"] = pd.to_numeric(clean_data["sq_meters"], downcast="integer")
```
